```python
mysize = (2000 / np.array(range(1,101))) # use 2000 to divide the array to disperse the numbers, the aim is to make  the numbers more condensed in the left side of the graph than the right side
```

```python
mysize.shape
```

```python
mysize
```

```python
chidist = []
```

```python
for i in range(0,len(mysize)):
    chidist.append(np.random.chisquare(100, mysize[i])) # i means the reference of the chidist
chidist
```

```python
type(chidist[0])
```

```python
chimeans = []
for i in range (0, len(mysize)):
    chimeans.append(chidist[i].mean())
```

```python
pl.scatter(mysize,chimeans)
```

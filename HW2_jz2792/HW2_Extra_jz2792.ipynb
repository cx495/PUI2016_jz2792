```python
df = df[['Date received','Complaint ID']]
df.columns = ['Date_received','Complaint_ID']
df.head(10)
```

```python
f = pd.value_counts(df.Date_received)
f= pd.DataFrame(f)
f['date'] = f.index
```

```python
f.reset_index(drop=True, inplace=True)
f.columns = ['No_Complaints','Date_received']
f.head(10)
```

```python
f_plot = f.plot(x='Date_received', y= 'No_Complaints',style='o', title='No. of Participants per Jurisdiction')
```
